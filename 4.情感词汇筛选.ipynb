### 删除词频小于5的词语

```python
with open("/Users/nxcy/Desktop/自然语言处理数据/分词结果txt和json/统计词频和词性还有分数.json", encoding="utf-8") as f:
    dict = json.load(f)
print(len(dict))
for i in list(dict.keys()):
    sum=0
    for j in dict[i]['词性'].keys():
        # print(dict[i]['词性'][j])
        sum=sum+dict[i]['词性'][j]
    if sum<5:
        print(i)
        dict.pop(i)
print(len(dict))
```
